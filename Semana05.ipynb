### 📋 Ejercicio 1: Autenticación con límite de intentos

**Instrucciones:**
Crea un programa que:
- Permita máximo 3 intentos de contraseña
- La contraseña correcta es "123456"
- Si es correcta: mostrar "Acceso permitido"
- Si se agotan intentos: mostrar "Cuenta bloqueada"

```python
# Solución:
intentos = 0
max_intentos = 3

while intentos < max_intentos:
    contraseña = input("Ingresa la contraseña: ")
    
    if contraseña == "123456":
        print("Acceso permitido")
        break
    else:
        intentos += 1
        print("Contraseña incorrecta")

if intentos == max_intentos:
    print("Cuenta bloqueada")
